## Feuille de route pour crée un PF efficient V1:

a) model du CAPM :  
un model assez pauvre dans son ensemble mais assez bon pour avoir une idée  
Décrit la rentabilité de l'actif comme la rentabilité de l'actif sans risque additionnée à une prime de risque marché (rentabilité du marché moins rentabilité de l'actif sans risque) pondérée par le beta de l'actif.

$$E(R_i) = R_f + \beta_{i}(E(R_m)-R_f)$$
$$\beta_{i} = \dfrac{Cov(R_i;R_m)}{Var(R_m)}$$

Où:  
$R_f$: taux sans risque (ex: obligation d'état, bon du trésor)  
$R_m$: représente le rendement de l'ensemble du marché (ou d'un indice de marché représentatif)  
$\beta$: rémuneration structuelle  
$\alpha$: surperformance non explicable par la chance  


b) model multi-factor:  
b.1) Model 3 facteurs de Fama-French  
$$ E(R_i) = R_f + \beta_i (R_m - R_f) + s_i \cdot SMB + h_i \cdot HML $$
Où: 

$
\begin{aligned}
E(R_i) &: \text{Rendement espéré de l'actif } i \\
R_m - R_f &: \text{Prime de risque du marché} \\
s_i &: \text{Sensibilité au facteur taille} \\
SMB &: \text{Small Minus Big (prime de taille)} \\
h_i &: \text{Sensibilité au facteur valeur} \\
HML &: \text{High Minus Low (prime de valeur)}
\end{aligned}
$

b.2) Model 5 facteurs  
$$E(R_i) = R_f + \beta_i (R_m - R_f) + s_i \cdot SMB + h_i \cdot HML + r_i \cdot RMW + c_i \cdot CMA $$

Avec :  

$
\begin{aligned}
RMW &: \text{Robust Minus Weak (prime de rentabilité)} \\
CMA &: \text{Conservative Minus Aggressive (prime d'investissement)}
\end{aligned}
$


## 1) La prime de Risque Pratique (ERP):
Taux réels, Term Prémium, Liquidité, Risk appetite

$µi$ : droite de regression  / esperence de rendement
$µi = Rf + \beta_{i} * ERP$  

## 2) Larchitecture du pf  
Markowitz & Modern pf theory  
Erreur fatal: la var d'un pf =/ moyenne des var 

pf 2 actifs :  
$\sigma_p^2 = w_1^2\sigma_1^2 + w_2^2\sigma_2^2 + 2w_1w_2\sigma_1\sigma_2\rho_{12}$  

Le dernier terme ($2w_1w_2\sigma_1\sigma_2\rho_{12}$) est crucial car il montre que le risque total n'est pas simplement la moyenne pondérée des risques individuels, mais dépend aussi de la corrélation entre les actifs. C'est ce qui permet la diversification du portefeuille.  

Où :

- $\sigma_p^2$ = Variance du portefeuille (mesure du risque total)  
- $w_1$ = Poids (proportion) de l'actif 1 dans le portefeuille  
- $w_2$ = Poids (proportion) de l'actif 2 dans le portefeuille  
- $\sigma_1^2$ = Variance de l'actif 1  
- $\sigma_2^2$ = Variance de l'actif 2  
- $\sigma_1$ = Écart-type (volatilité) de l'actif 1  
- $\sigma_2$ = Écart-type (volatilité) de l'actif 2  
- $\rho_{12}$ = Coefficient de corrélation linéaire entre les rendements de l'actif 1 et de l'actif 2  

Et avec les contraintes :  

$w_1 + w_2 = 1$ (la somme des poids est égale à 100\% du portefeuille)  

$-1 \leq \rho_{12} \leq 1$ (le coefficient de corrélation est toujours compris entre -1 et +1)  La clé: $p<1$

## 4) La Fonction D'utilité

$$
\max_{W} \left( \underbrace{w^T \mu}_{\text{Rendement espéré}} - \underbrace{\frac{\lambda}{2} w^T \Sigma w}_{\text{Pénalité de risque}} \right)
$$

Avec :

- $w$ : Vecteur des poids des actifs
- $\mu$ : Vecteur des rendements espérés
- $\Sigma$ : Matrice de variance-covariance
- $\lambda \geq 0$ : Coefficient d'aversion au risque
- $w^T \mu$ : Rendement espéré du portefeuille
- $w^T \Sigma w$ : Variance du portefeuille ($\sigma_p^2$)  

a) Solution analytique

$$
w^* = \frac{1}{\lambda} \Sigma^{-1} \mu
$$

Où :

- $w^*$ : Vecteur des **poids optimaux** du portefeuille
- $\Sigma^{-1}$ : Inverse de la matrice de variance-covariance
- $\mu$ : Vecteur des rendements espérés des actifs
- $\lambda \geq 0$ : Coefficient d'aversion au risque
- $\frac{1}{\lambda}$ : Tolérance au risque (inverse de l'aversion)

## 7) Le Solver: l'optimisation 

Problème d'optimisation convexe (QP)  
exemple de contrainte:  

$$\min_{w} \frac{1}{2} w^T \Sigma w - \kappa \mu^T w $$


Sous contraintes :

- $ \Sigma w_i = 1$ (contrainte de budget)
- $w_i \geq 0, \quad \forall i$ (pas de vente à découvert)
- cardinalité: Max $\kappa$ actifs 

Où:  
- $w = (w_1, \dots, w_n)^T$ : Vecteur des poids des actifs
- $\Sigma$ : Matrice de variance-covariance
- $\mu$ : Vecteur des rendements espérés
- $\kappa \geq 0$ : Paramètre de préférence pour le rendement
